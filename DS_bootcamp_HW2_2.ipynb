```python
# Replace missing values in Min.Price and Max.Price columns with their respective mean (check documentation).

import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
df.head(10)
```

```python
Min_Price = "Min.Price"
Min_data = df[Min_Price]

row_number = df.shape[0]

Mean_Min = Min_data.mean()
Mean_Min = round(Mean_Min,1)
df.head()

for i in range (row_number):
    if (pd.isna(Min_data[i])) :
        df[Min_Price][i] = Mean_Min

df
```

```python
Max_Price = "Max.Price"
Max_data = df[Max_Price]

row_number = df.shape[0]

Mean_Max = Max_data.mean()
Mean_Max = round(Mean_Max,1)
df.head()

for i in range (row_number):
    if (pd.isna(Max_data[i])) :
        df[Max_Price][i] = Mean_Max

df
```
